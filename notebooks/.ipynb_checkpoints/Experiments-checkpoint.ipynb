```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns
sns.set()
from utils import *
import scipy.misc

```

```python
train_colours = sns.color_palette('bright', 10)#sns.dark_palette("blue", 10)#sns.color_palette("hls", 10)
test_colours = train_colours#sns.dark_palette("blue", 10)#sns.color_palette("hls", 10)#sns.cubehelix_palette(6, start=.5, rot=-.75, reverse=True)

sns.palplot(train_colours)
sns.palplot(test_colours)
```

```python
results = {}
model = 'cnn'
loss_ylims = {'cnn': [0,1.5], 'preact_resnet':[0,1]}
loss_xlims = {'cnn': [0,200], 'preact_resnet':[0,400]}
acc_ylims = {'cnn': [0.65,1.005], 'preact_resnet':[0.78,1.005]}
acc_xlims = {'cnn': [0,200], 'preact_resnet':[0,400]}
iterations_processed = range(3)
log_folder = 'log'
for iteration in iterations_processed:
#     if iteration == 9:
#         results[iteration] = read_results('../{}/perepochcal-{}/summary_logs/result_summary_statistics.csv'.format(log_folder, model))
#     else:
    results[iteration] = read_results('../{}/iter-{}-{}/summary_logs/result_summary_statistics.csv'.format(log_folder, iteration, model))
```

```python
fig = plt.figure(figsize=(18, 9))

ax = fig.add_subplot(121)
for iteration in iterations_processed:
    epochs = results[iteration]['epoch']
    train_loss = results[iteration]['train_loss']    
    test_loss = results[iteration]['test_loss']
    ax.plot(epochs, train_loss, color=train_colours[iteration], 
            label='{}-{} ({:0.4f})'.format(model, iteration, np.min(train_loss)))
    ax.plot(epochs, test_loss, color=test_colours[iteration], linestyle='--', 
            label='{}-{} Test ({:0.4f})'.format(model, iteration, np.min(test_loss)))

ax.set_ylim(loss_ylims[model])
ax.set_xlim(loss_xlims[model])
ax.set_xlabel('Epochs')
ax.set_ylabel('Loss')
ax.set_title('{} losses'.format(model))
ax.legend(frameon=False)


ax = fig.add_subplot(122)
for iteration in iterations_processed:
    epochs = results[iteration]['epoch']
    train_acc = results[iteration]['train_acc']    
    test_acc = results[iteration]['test_acc']
    ax.plot(epochs, train_acc, color=train_colours[iteration], 
            label='{}-{} ({:0.4f})'.format(model, iteration, 100 * np.max(train_acc)))
    ax.plot(epochs, test_acc, color=test_colours[iteration], linestyle='--', 
            label='{}-{} Test ({:0.4f})'.format(model, iteration, 100 * np.max(test_acc)))

ax.set_ylim(acc_ylims[model])
ax.set_xlim(acc_xlims[model])
ax.set_xlabel('Epochs')
ax.set_ylabel('Acc')
ax.set_title('{} accuracies'.format(model))
ax.legend(frameon=False)

fig.tight_layout()
```
